```python
df['Memory'] = df['Memory'].astype(str)
df['Memory'] = df['Memory'].replace(r'\.0', '', regex = True)
df['Memory'] = df['Memory'].str.replace('GB', '', regex = False)
df['Memory'] = df['Memory'].str.replace('TB', '000', regex = False)

new = df['Memory'].str.split('+', n=1, expand=True)

df['first'] = new[0].str.strip()
df['second'] = new[1].str.strip() if new.shape[1] > 1 else '0'

df['layer1HDD'] = df['first'].apply(lambda x: 1 if 'HDD' in x else 0)
df['layer1SSD'] = df['first'].apply(lambda x: 1 if 'SSD' in x else 0)
df['layer1FlashStorage'] = df['first'].apply(lambda x: 1 if 'Flash Storage' in x else 0)
df['layer1Hybrid'] = df['first'].apply(lambda x: 1 if 'Hybrid' in x else 0)

df['first'] = df['first'].str.replace(r'\D', '', regex=True)

df['second'] = new[1].fillna('0')
df['second'] = df['second'].str.strip()

df['layer2HDD'] = df['second'].apply(lambda x: 1 if 'HDD' in x else 0)
df['layer2SSD'] = df['second'].apply(lambda x: 1 if 'SSD' in x else 0)
df['layer2FlashStorage'] = df['second'].apply(lambda x: 1 if 'Flash Storage' in x else 0)
df['layer2Hybrid'] = df['second'].apply(lambda x: 1 if 'Hybrid' in x else 0)

df['second'] = df['second'].str.replace(r'\D', '', regex=True)

df['first'] = df['first'].astype(int)
df['second'] = df['second'].astype(int)

df['HDD'] = (df['first'] * df['layer1HDD'] + df['second'] * df['layer2HDD'])
df['SSD'] = (df['first'] * df['layer1SSD'] + df['second'] * df['layer2SSD'])
df['FlashStorage'] = (df['first'] * df['layer1FlashStorage'] + df['second'] * df['layer2FlashStorage'])
df['Hybrid'] = (df['first'] * df['layer1Hybrid'] + df['second'] * df['layer2Hybrid'])

df.drop(columns=['first', 'second', 'layer1HDD', 'layer1SSD', 'layer1FlashStorage', 'layer1Hybrid', 'layer2HDD', 'layer2SSD', 'layer2FlashStorage', 'layer2Hybrid'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.15, random_state=2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.ensemble import VotingRegressor,StackingRegressor
import joblib
import pickle
```

```python
models = {
    'LinearRegression': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', LinearRegression())
    ]),
    'Ridge': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', Ridge(alpha=14))
    ]),
    'Lasso': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', Lasso(alpha=0.001))
    ]),
    'KNeighborsRegressor': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', KNeighborsRegressor(n_neighbors=3))
    ]),
    'DecisionTreeRegressor': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', DecisionTreeRegressor(max_depth=8))
    ]),
    'SVR': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', SVR(kernel='rbf', C=10000, epsilon=0.16))
    ]),
    'RandomForestRegressor': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', RandomForestRegressor(n_estimators=100, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15))
    ]),
    'ExtraTreesRegressor': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', ExtraTreesRegressor(n_estimators=100, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15, bootstrap=True))
    ]),
    'AdaBoostRegressor': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', AdaBoostRegressor(n_estimators=15, learning_rate=1.0))
    ]),
    'GradientBoostingRegressor': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', GradientBoostingRegressor(n_estimators=500))
    ]),
    'XGBRegressor': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', XGBRegressor(n_estimators=45, max_depth=5, learning_rate=0.5))
    ]),
    'VotingRegressor': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', VotingRegressor([
            ('rf', RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15)),
            ('gbdt', GradientBoostingRegressor(n_estimators=100, max_features=0.5)),
            ('xgb', XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5)),
            ('et', ExtraTreesRegressor(n_estimators=100, random_state=3, max_samples=0.5, max_features=0.75, max_depth=10, bootstrap=True))
        ], weights=[5, 1, 1, 1]))
    ]),
    'StackingRegressor': Pipeline([
        ('step1', ColumnTransformer([
            ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
        ], remainder='passthrough')),
        ('step2', StackingRegressor(
            estimators=[
                ('rf', RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15)),
                ('gbdt', GradientBoostingRegressor(n_estimators=100, max_features=0.5)),
                ('xgb', XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5))
            ],
            final_estimator=Ridge(alpha=100)
        ))
    ])
}

best_model = None
best_r2_score = float('-inf')

metrics_data = []

# Fitting and evaluating models
for model_name, pipe in models.items():
    X_train_array = X_train.to_numpy()
    X_test_array = X_test.to_numpy()
    
    pipe.fit(X_train_array, y_train)
    y_pred = pipe.predict(X_test_array)
    
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    # Appending metrics and its performances in dictionary
    metrics_data.append([r2, mae, mse, rmse]) 

    # Saving the best model based on R2 score
    if model_name == 'RandomForestRegressor':
        best_model = pipe

# Saving the best model
if best_model:
    joblib.dump(best_model, 'RandomForestRegressor_pipeline.joblib')

# Saving the dataframe
with open('dataframe.pkl', 'wb') as f:
    pickle.dump(df, f)
```

```python
# creating dataframe
df_metrics = pd.DataFrame(metrics_data, columns=['R', 'MAE', 'MSE', 'RMSE'], index=models.keys())
```

```python
# Normalizing the dataframe
df_normalized = (df_metrics - df_metrics.mean()) / df_metrics.std()
Z = linkage(df_normalized, 'ward')

# Create a dendrogram
plt.figure(figsize=(19, 6))
dendrogram_data = dendrogram(Z, labels=list(models.keys()), leaf_rotation=45, leaf_font_size=12)
plt.title('Dendrogram of Model Performance')
plt.xlabel('Models')
plt.ylabel('Distance')
plt.show
```
